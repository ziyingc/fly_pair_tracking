```python
from os import listdir
from os.path import isfile, join
folder_path = '../../FlyBehaviorSet/DLC_models/Outliers/Outlier_Frames_05_08_24_exp1_top_view'
directories = [item for item in os.listdir(folder_path) if os.path.isdir(os.path.join(folder_path, item))]
onlyfiles = []
for mypath in directories:
    x = listdir(folder_path+'/'+mypath)
    for img in x:
        onlyfiles.append(folder_path+'/'+mypath+'/'+img)
#         print(folder_path+'/'+mypath+'/'+img)

outlier_video = folder_path+'/Outlier_Frames_05_08_24_exp1_top_view.avi'
frame_rate = 60
outlier_frames_video(outlier_video, onlyfiles, frame_rate)
```

```python
import cv2
import numpy as np
def outlier_frames_video(output_file, images, frame_rate):
    fourcc = cv2.VideoWriter_fourcc(*'FFV1')

        
    frame = cv2.imread(images[0])
    
    height, width, layers = frame.shape
    out = cv2.VideoWriter(output_file, fourcc, frame_rate, (width, height))

    for image in images:
        frame = cv2.imread(image)
        out.write(frame)
    out.release()
    cv2.destroyAllWindows()
    print("outlier vidoe created")
```
